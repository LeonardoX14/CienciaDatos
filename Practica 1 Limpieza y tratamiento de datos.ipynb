```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
import numpy as np
```

```python
autos= pd.read_csv("autos.csv")
autos.head()
```

```python
autos.info()
```

```python
autos.replace("?", np.nan, inplace=True)
```

```python
autos.head()
```

```python
autos.info()
```

```python
autos["normalized-losses"].isnull().value_counts()
```

```python
promedio_normalized_losses = autos["normalized-losses"].astype("float").mean(axis=0)
print(promedio_normalized_losses)
```

```python
autos["normalized-losses"].replace(np.nan, promedio_normalized_losses, inplace=True)
autos.head()
```

```python
##REMPLAZO DE NAN EN BORE
avg_bore = autos["bore"].astype("float").mean(axis=0)
print(avg_bore)
```

```python
autos["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
## REMPLAZO DE NAN EN STROKE
avg_stroke = autos["stroke"].astype("float").mean(axis=0)
print(avg_bore)
autos["stroke"].replace(np.nan, avg_bore, inplace= True)
```

```python
##REMPLAZO DE NAN EN HORSEPOWER
avg_horsepower= autos["horsepower"].astype("float").mean(axis=0)
print(avg_horsepower)
autos["horsepower"].replace(np.nan, avg_horsepower, inplace= True)
```

```python
##REMPLAZO DE NAN EN PEAK-RPM

avg_peak_rpm= autos["peak-rpm"].astype("float").mean(axis=0)
print(avg_peak_rpm)
autos["peak-rpm"].replace(np.nan, avg_peak_rpm, inplace=True)
```

```python
autos["num-of-doors"].value_counts()
```

```python
##REMPLAZAMOS POR EL NUMERO MAYOR DE PUERTAS
autos["num-of-doors"].replace(np.nan, "four", inplace = True)
autos.info()
```

```python
##Faltante es el del Precio
avg_price= autos["price"].astype("float").mean(axis=0)
print(avg_price)
autos["price"].replace(np.nan, avg_price, inplace= True)
autos.info()
```

```python
##5 CORREGIR EL FORMATO DE DATOS
autos.dtypes
```

```python
autos[["bore","stroke"]]= autos[["bore","stroke"]].astype("float")
autos[["normalized-losses"]]= autos[["normalized-losses"]].astype("int")
autos[["price"]]= autos[["price"]].astype("float")
autos[["peak-rpm"]]= autos[["peak-rpm"]].astype("float")
```

```python
autos.dtypes
```

```python
##6 Estandarizacion de los datos
## Nueva columna de Conversion de las unidades

autos["city-L/100km"]= 235/autos["city-mpg"]
```

```python
##Revisamos los datos ya transformados
autos.head()
```

```python
autos["highway-mpg"]= 235/autos["highway-mpg"]

##Renombramos la columna anterior
autos.rename(columns={"highway-mpg":"highway-L/100km"},inplace= True)

##Revisamos los datos nuevamente 
autos.head()
```

```python
##7. NORMALIZACION DE LOS DATOS

#Primero visualizamos las columnas a normalisar
autos[["length","width","height"]].head()
```

```python
## Intercambiamos el valor original por el valor original / valor maximo

autos["length"]= autos["length"]/ autos["length"].max()
autos["width"]= autos["width"]/ autos["width"].max()
autos["height"]= autos["height"]/ autos["height"].max()
##Verificacion Rapida
autos[["length","width","height"]].head()
```

```python
##8 VARIABLES DUMMY O INDICADORES

autos["fuel-type"].value_counts()
```

```python
##Codificamos las variables 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#Agregamos una nueva columna
autos["fuel-type-code"]= le.fit_transform(autos["fuel-type"])
```

```python
autos.head()
```

```python
##Verificamos la columna que se agrego
autos["fuel-type-code"].value_counts()
```

```python
##Convertir variable categorica en ficticia 

variable_dummy1= pd.get_dummies(autos["fuel-type"])
variable_dummy1.head()
```

```python
#Combinar el dataframe 
autos= pd.concat([autos, variable_dummy1], axis=1)
autos.head(10)
```

```python
##9 DISCRETIZACION DE VARIABLES
autos["horsepower"] = autos["horsepower"].astype(int, copy=True)
plt.pyplot.hist(autos["horsepower"])

#ESTABLECEMOS ETIQUETAS
plt.pyplot.xlabel("Caballos de Fuerza")
plt.pyplot.ylabel("Cantidad")
plt.pyplot.title("Contenedores horsepower")
```

```python
##ESTABLECEMOS EL VALOR DE BINS IGUAL A 3
plt.pyplot.hist(autos["horsepower"],bins = 3)

plt.pyplot.xlabel("Caballos de Fuerza")
plt.pyplot.ylabel("Cantidad")
plt.pyplot.title("Contenedores horsepower")
```

```python
##CONTENEDORES DEL RANGO DE VALORES
bins = np.linspace( min(autos["horsepower"]), max(autos["horsepower"]), 4)
bins
```

```python
##DEFINIMOS LOS NOMBRES DE GRUPO.
grupos = ["Low","Medium","High"]
```

```python
##DEFNIMOS UNA NUEVA COLUMNA
autos["horsepower-binned"]=pd.cut(autos["horsepower"],bins,labels=grupos,include_lowest=True)

##VERIFICAMOS
autos[["horsepower","horsepower-binned"]].head(10)
```

```python
##VERIFICAMOS LO QUE SON LAS CARACTERISTICAS DE LA NUEVA COLUMNA
autos["horsepower-binned"].value_counts()
```

```python
pyplot.bar(grupos, autos["horsepower-binned"].value_counts())

plt.pyplot.xlabel("Caballos de Fuerza")
plt.pyplot.ylabel("Cantidad")
plt.pyplot.title("Contenedores horsepower")
```
